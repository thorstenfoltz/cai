```python
import os
from dotenv import load_dotenv
import logging
logging.basicConfig(
    format="%(asctime)s.%(msecs)03d %(levelname)s %(name)s: %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)
log = logging.getLogger(__name__)
log.setLevel(logging.INFO)
```

```python

key = "OpenAIS"

def load_key(key) -> str:
    log.info(f"Load access token to {key}")
    try:
        load_dotenv()
        api_key = os.getenv(key)
        if not api_key:
            raise ValueError(f"Key '{key}' not found in environment variables.")
        log.info(f"Token to {key} successfully loaded")
        return api_key
    except Exception as e:
        log.error(f"Failed to load token for {key}: {e}")
        raise
        
    
```
